```python
def closestPairs(distances,clusters):    n = len(distances)    spacing = infinity    clustersToMerge = (0,0)    for i in xrange(n):        for j in xrange(i+1,n):            if clusters[i] != clusters[j] and distances[i][j] < spacing:                spacing = distances[i][j]                clustersToMerge = [clusters[i],clusters[j]]    return spacing,clustersToMerge
```

```python
def mergeClusters(distances,k):    n = len(distances)    clusters = range(n)    for i in xrange(n-k):        i,clustersToMerge = closestPairs(distances,clusters)        for j in xrange(n):            if clusters[j] == clustersToMerge[1]:                clusters[j] = clustersToMerge[0]    return clusters
```
